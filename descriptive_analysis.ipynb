Menghitung penyebaran data

```python
range_total_char = df.total_char.max() - df.total_char.min()
range_total_char
```

```python
#quartile dan interquartile range

p0 = df.total_char.min() #nilai minimum
p100 = df.total_char.max() #nilai maksimum

#cari quartile 1
q1 = df.total_char.quantile(0.25)

#cari quartile 2
q2 = df.total_char.quantile(0.5)

#cari quartile 3
q3 = df.total_char.quantile(0.75)

#cari IQR
iqr = q3 - q1

#batas bawah
lower_limit = q1 - 1.5 * iqr

#batas atas
upper_limit = q3 + 1.5 * iqr
```

```python
print("batas bawah 'total_char' : ", lower_limit)
print("nilai minimum: ",p0)

if lower_limit<p0:
  print("tidak ada outlier dari sisi bawah")
else:
  print("ada outlier dari sisi bawah")

print("batas atas 'total_char' : ", upper_limit)
print("nilai maksimum: ",p100)

if upper_limit>p100:
  print("tidak ada outlier dari sisi atas")
else:
  print("ada outlier dari sisi atas")
```

```python
range_total_word = df.total_word.max() - df.total_word.min()
range_total_word
```

```python
#quartile dan interquartile range

p0w = df.total_word.min() #nilai minimum
p100w = df.total_word.max() #nilai maksimum

#cari quartile 1
q1w = df.total_word.quantile(0.25)

#cari quartile 2
q2w = df.total_word.quantile(0.5)

#cari quartile 3
q3w = df.total_word.quantile(0.75)

#cari IQR
iqrw = q3 - q1

#batas bawah
lower_limit_word = q1w - 1.5 * iqrw

#batas atas
upper_limit_word = q3w + 1.5 * iqrw
```

```python
print("batas bawah 'total_word' : ", lower_limit_word)
print("nilai minimum: ",p0w)

if lower_limit_word<p0w:
  print("tidak ada outlier dari sisi bawah")
else:
  print("ada outlier dari sisi bawah")

print("batas atas 'total_word' : ", upper_limit_word)
print("nilai maksimum: ",p100w)

if upper_limit_word>p100w:
  print("tidak ada outlier dari sisi atas")
else:
  print("ada outlier dari sisi atas")
```
